# Importing Modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import tree
```

# Data Extracting

```python
df=pd.read_csv('iris.csv',index_col=0)
```

```python
df.head()
```

```python
X=df.iloc[:,[0,1,2,3]]
```

```python
print(X)
```

```python
y=df.iloc[:,-1]
```

```python
y
```

# Feature Scaling

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
```

```python
y
```

# Generating data for Training and test datasets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
X_train
```

```python
y_train
```

# Generating the Decision tree model using Entropy as criteria

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

# Getting Confusion Matrix and Accuracy of Model(Entropy)

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(cm,annot=True,xticklabels=['Sesota','versicolor','virginica'],yticklabels=['Sesota','versicolor','verginica'])
plt.xlabel('Prediction Label')
plt.ylabel('True Value')
plt.title('Confusion Matrix Without Normalization')
```
